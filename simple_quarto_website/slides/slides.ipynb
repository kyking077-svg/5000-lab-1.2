## Spring {background-color=#DFFFE0}

::: {.incremental}
- Most crops grow quickly, but sell for less as well
- All crops are single-season crops (they die at the beginning of the next season)
- Best season for quick turnarounds and flexible plant beds
![](images/spring.jpg)
:::
## Summer {background-color=#B0E0E6}
::: {.incremental}
- Season with the highest number of high-profit plants
- Corn and Wheat can be planted in Summer and grow through Fall
- Best season for profitable single instance harvests
![](images/summer.jpg)
:::

## Fall {background-color=#F5CBA7}
::: {.incremental}
- Has the most multi-harvest crops. If you plant them at the beginning of the season, they produce throughout
- Corn and Wheat from Summer provide early season income
- Best season for consistent earnings and minimal re-plantings
- Has my favorite songs in the soundtrack, which [@galloway2019soundwalking] says adds to my experience.
![](images/fall.jpg)
:::

## Winter {background-color=#CFEFFF}
::: {.incremental}
- Only four crops grow in Winter
- Best time to visit the Mines and gather resources
- Best season for community interaction because you are less busy on the farm
![](images/winter.jpg)
:::


## Benefits of Stardew Valley {background-color=#FFE4E1}
This chart shows how much my stress decreases as I play Stardew Valley

```python
import matplotlib.pyplot as plt
import numpy as np

# Data: hours played vs stress
hours_played = np.arange(1, 11)
stress_level = 100 * np.exp(-0.2 * hours_played)  # exponential decay

# Line plot
plt.figure(figsize=(8,5))
plt.plot(hours_played, stress_level, marker='o', color='skyblue', linewidth=2)
plt.title("Stress Decline vs. Time Playing Stardew Valley")
plt.xlabel("Hours Played")
plt.ylabel("Stress Level")
plt.ylim(0, 100)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
